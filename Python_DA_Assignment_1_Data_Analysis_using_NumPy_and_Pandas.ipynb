Numpy Array Operations:

```python
#2. Inspection and Properties:
print(temperatures_w1)
# Data Types
print("Data Types",temperatures_w1.dtype)
# Shape
print("Shape",temperatures_w1.shape)
#Number of Elements
print("Number of Elements",temperatures_w1.size)
```

Pandas Series:

**1. Creating Pandas Series:**

```python
import pandas as pd
```

```python
# Pandas Series
marks=pd.Series([95, 92, 89, 85, 80],
                index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5'])
print(marks)
print(type(marks))
```

**2. Indexing and Slicing:**

```python
# Use the integer index position to access the mark of the 1st rank student.
print(" 1st rank student \n",marks.iloc[0])
# Use the loc accessor to retrieve the marks of the top 3 ranks by specifying their inde labels.
print("top 3 ranks \n",marks.iloc[0:3])
# Use the iloc accessor to retrieve the mark of the 3rd-ranked student.
print("retrieve the mark of the 3rd-ranked student \n",marks.iloc[2])
# Apply a boolean mask to filter the Series and retrieve the ranks where the marks ar greater than 90.
print("Ranks with marks > 90:\n", marks[marks > 90])
```

**3. Manipulating Series:**

```python
# Modify the mark of the student with the 1st rank to 100.
marks["Rank1"]=100
print("Modifying data",marks)
# Remove the entry corresponding to the last rank from the Series.


marks = marks.drop("Rank5")
print("droping data",marks)
# Compute the CGPA by dividing each mark in the Series by 10.
CGPA=marks/10
print("CGPA",CGPA)
```

**Pandas DataFrame:

1. Creating Pandas DataFrame:**

```python
transactions={
    "TransactionID" : [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "ProductCategory" : ['Electronics', 'Clothing', 'Electronics', 'Furniture', 'Clothing','Electronics', 'Furniture', 'Clothing', 'Furniture', 'Electronics'],
    "Region" : ['North', 'South', 'North', 'East', 'West', 'North', 'East', 'West', 'South','North'],
    "Amount" : [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]

}
print(transactions)
df=pd.DataFrame(transactions)
```

2**. Data Exploration:**

```python
#Display the transactions DataFrame and its basic information, including head, tail,shape, column names, and data types.
print("Head:", df.head())
print("Tail", df.tail())
print("Shape ", df.shape)
print("Column names", df.columns)
# Display only the 'ProductCategory' and 'Amount' columns.
print("ProductCategory & Amount:\n", df[["ProductCategory", "Amount"]])
# Retrieve the last 3 columns of the DataFrame using iloc or loc
print("Last 3 columns:\n", df.iloc[:, -3:])
# Filter rows where the 'Region' is 'North' and 'Amount' is greater than 200.

print("Filter Region and Amount greater than 200:\n", df[(df["Region"] == "North") & (df["Amount"] > 200)])
# Find the value counts for the 'ProductCategory' column.
print("Value_Counts of Product Category",df["ProductCategory"].value_counts())
# Find the unique values in the 'Region' column.
print("Unique Region : ",df["Region"].unique())
# Group by 'Region' and find the mean amount for each region.
print("Region_Mean Using Groupby",df.groupby("Region")["Amount"].mean())
```
